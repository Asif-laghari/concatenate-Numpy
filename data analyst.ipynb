# Loading it into pandas data frame Csv files

```python
import pandas as pd

sample_df = pd.read_csv("examples/sample.csv")
print(sample_df)


# we can change column name while loading data
sample_df = pd.read_csv("examples/sample.csv", names = ["id", "name", "ch", "ph", "eng", "math"])
print(sample_df) # when we add column names it will be convert into data culumns names

# we should do skiprows
sample_df = pd.read_csv("examples/sample.csv", skiprows=1, names = ["id", "name", "ch", "ph", "eng", "math"])
print(sample_df)
```
